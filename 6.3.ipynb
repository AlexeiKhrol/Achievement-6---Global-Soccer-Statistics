```python
#3 Import the libraries you need for your analysis (e.g., Geopandas and Geoplot).
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import os
import geopandas as gpd
import geoplot
```

```python
path = r'C:\Users\alexk\Achievement 6'
```

```python

#4 Find a shapefile containing location data that corresponds to the location data in your student project data.
map = gpd.read_file(os.path.join(path, 'ne_50m_admin_0_countries.shp'))
```

```python
map.head()
```

```python
map.rename(columns = {'country' : 'Country'}, inplace = True)
```

```python
map.head()
```

```python
df = pd.read_csv(os.path.join(path,'att.csv'))
```

```python
df
```

```python
#5 Wrangle your project data to meet the needs of your analysis.
df['Highest Attendance'] = df.groupby(['Country'])['Attendance'].transform(np.max)
```

```python
df.head()
```

```python
#6 Merge the two data files into a single GeoDataFrame. 
df_merged = map.merge(df, on = 'Country', how = 'inner', indicator = True)
```

```python
df_merged
```

```python
#8 Create a choropleth map using a variable of your choice.
gp = geoplot.choropleth(
    df_merged, hue = df_merged['Highest Attendance'], scheme = None,
    cmap='Reds', figsize=(15, 15)
)
```
